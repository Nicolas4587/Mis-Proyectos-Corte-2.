**Titulo del proyecto:**

Evaluación del ruido ambiental en el campus de la Universidad de Santander con un módulo electrónico de bajo costo en tiempo real.

**Nombre de la universidad:** Universidad de Santander (UDES)


**Resumen**
Este proyecto realizo un dispositivo que genera datos en decibeles del sonido por medio de un celular mediante bluetooth midiendo constantemente en tiempo real el ruido del ambiente basados en la resolución 627 del 2006, esta resolución les permitio conocer que en un espacio cerrado el ruido promedio sano es de 67 a 70 decibeles y en un espacio abierto lo máximo debe ser 80 decibeles.Los integrantes de este proyecto hicieron las pruebas en tres puntos de la Universidad de Santander en el horario de 5pm a 7pm,lo que dio como resultado que se encuentra los decibeles dentro del rango de la resolución 627.
El objetivo de este proyecto es evitar la contaminación auditiva e inluso los problemas respecto al ruido como proteger a las aves, al ecosistema y nosotros mismos.

**Nombre de la linea de estudio:** Calidad de aire y cambio climátio.


**Grupo de investigación al que esta asociado el proyecto:** GAIA Grupo ambiental de investigación aplicada.

**Titulo del proyecto:**

Expansión de células madres mesenquimales en condiciones hipóxicas en medio no xenogénico suplementado con plasma pobre en plaquetas.

**Nombre de la universidad:** Universidad Autónoma de Bucaramanga (UNAB)


**Resumen**
El objetivo de este proyecto es evaluar el efecto de las condiciones hipóxicas y medio no xenogénico en la expansión de MSC mediante la elaboración de cultivos in vitro en condiciones hipóxicas suplementando con plasma pobre en plaquetas(PPC) que permite la expansión del MSC con el fin de usarse en la medicina regenerativa en el desarrollo de métodos para la regeneración de tejidos.Las celulas madres son las que mas han dado resultados en el campo de la medicina regenerativa gracias a sus propiedades, a lo largo del estudio se identifico un problema, que actualmente en los hospitales y en el ambiente investigativo estan usando como medios de cultivo el suero fetal de bovino con el problema que puede tener algunas proteinas que puede generar efectos negativos al paciente, por lo que, una forma de contrarestar este riesgo es usando el plasma pobres en plaquetas que se puede extraer del torrente sanguineo.


**Nombre de la linea de estudio:** Ingeniería de tejidos y terapia celular de células madres


**Grupo de investigación al que esta asociado el proyecto:** Medicina Regenerativa
